```python
import numpy as np
import pandas as pd
import os
import seaborn as sns
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
import shutil

from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
def dist_fix(datagen, df_name, image_count, filename, trainpath):
    
    for i in range(5):
        if image_count[i] > 0:
            df_subset = df_name[df_name['annotation'] == i]

            for k in range(image_count[i]):
                make_it = datagen.flow_from_dataframe(df_subset,x_col='file_name', y_col='annotation_string',
                                                                directory = trainpath, 
                                                                target_size = (480,640), # some reason needs it inverted 
                                                                shuffle=True, batch_size=1,
                                                                save_to_dir=filename[i], save_prefix='aug',
                                                                save_format='jpg')

                make_it.next();
            
            
def move_files(base_dir, nested_top_folder_name, df):
    for i in range(5):  # iterate over data frame for each classifier
        unique_file_list = df[df.annotation == i].file_name.tolist()  # retrievs files by classification
        
        # Source path  
        
#         print(base_dir)
        
        # Destination Path
        class_folder_name = os.path.join(nested_top_folder_name, str(i))
        destination = os.path.join(my_dir, class_folder_name)
#         print(destination)
        
        for a_file in unique_file_list:
           # Source path
            file_source = os.path.join(base_dir, a_file)
            dest = shutil.copy(file_source, destination)
```

```python
my_dir = os.getcwd()  # grabs current work dir
all_file_paths = os.listdir()  # grabs all items in current work dir

FOLDER_TO_ACCESS = 'traindata_pp'
data_path = [data for data in all_file_paths if FOLDER_TO_ACCESS in data.lower()]  # searches for data folder and retrieves
train_folder_path = os.path.join(my_dir, data_path[0])  # joins data folder and current work dir to get picture path
all_classifications = os.listdir(train_folder_path)  # list of classification fodlers
```

```python
df_Train = pd.read_csv('TrainAnnotations.csv')
df_Train['annotation_string']=df_Train['annotation'].astype('str') # for KERAS

image_counts = []

for i in range(5):
    val_count = df_Train.annotation.value_counts(normalize=False, sort=False).iloc[i]
    image_counts.append(val_count)
    print(f'Classification {i} = {val_count} images')

max_image_count = max(image_counts)
image_counts = np.asarray(image_counts)
images_to_make = (max_image_count - image_counts)

images_to_make
```

Make Structured Folder Path for files

```python
file_name = []
base_folder_name = 'blob'
for i in range(5):
    try:  
        file_name.append(os.path.join(base_folder_name, str(i)))
        os.makedirs(file_name[i])  
    except OSError as error:  
        print(error)     
```

```python
move_files(train_folder_path, base_folder_name,  df_Train)
```

```python
datagen_processor = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.1,
                                   zoom_range=0,
                                   horizontal_flip=True,
                                   rotation_range=10,
                                   fill_mode = 'wrap')

dist_fix(datagen_processor, df_Train, images_to_make, file_name, train_folder_path)
# for i in range(5):
#     if images_to_make[i] > 0:
#         df_subset = df_Train[df_Train['annotation'] == i]
        
#         for k in range(images_to_make[i]):
#             make_it = datagen_processor.flow_from_dataframe(df_subset,x_col='file_name', y_col='annotation_string',
#                                                             directory = train_folder_path, # some reason needs it inverted 
#                                                             target_size = (480,640),
#                                                             shuffle=True, batch_size=1,
#                                                             save_to_dir=file_name[i], save_prefix='aug',
#                                                             save_format='jpg')

#             make_it.next()
```
