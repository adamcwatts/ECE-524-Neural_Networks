```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
sns.set_style("darkgrid")
sns.set_context("talk")
```

```python
sns.set_style("darkgrid") # darkgrid style
sns.set_context("talk")  # font size
sns.set_style('darkgrid',{'xtick.bottom': True, 'ytick.left': True}) # customize like a pro
sns.axes_style()

SMALL_SIZE = 20
MEDIUM_SIZE = 22
BIGGER_SIZE = 24
plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
df_hist = pd.read_csv('model_history_color_hist_50drop_corrected.csv', index_col=False)
df_hist
```

```python
fig, axe = plt.subplots(figsize=(6, 6))
plt.plot(np.arange(1, 19),df_hist['accuracy'], label='Training')
plt.plot(np.arange(1, 19),df_hist['val_accuracy'], label='Validation')
plt.xlim(1, 18)
tik = np.arange(2, 19, 2)
tik = np.insert(tik, 0, 1)
plt.xticks(tik)
plt.xlabel('Epochs', labelpad=10)
plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout()
plt.savefig('accuracy.png')



fig, axe = plt.subplots(figsize=(6, 6))
plt.plot(np.arange(1, 19),df_hist['loss'], label='Training')
plt.plot(np.arange(1, 19),df_hist['val_loss'], label='Validation')
plt.xlim(1, 18)
tik = np.arange(2, 19, 2)
tik = np.insert(tik, 0, 1)
plt.xticks(tik)
plt.xlabel('Epochs', labelpad=10)
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.savefig('Loss.png')
```

```python
fig, axe = plt.subplots(figsize=(6, 6))
df_y = pd.read_csv('TrainAnnotations.csv', index_col=False)
bin_val = np.arange(0, 6, 1) - 0.5
sns.distplot(df_y['annotation'], kde=False, bins=bin_val)
# plt.xlim(-0.1,4.1)
plt.xticks(np.arange(0,5, 1))
plt.ylabel("Training Set: Bin Count", labelpad=10)
plt.xlabel('Classification')
plt.tight_layout()
plt.savefig('training_set_histogram.png')
```

```python
fig, axe = plt.subplots(figsize=(6, 6))
df_pred = pd.read_csv('ModelPredictioncolor_hist_50drop_corrected.csv', index_col=False)
bin_val = np.arange(0, 6, 1) - 0.5
sns.distplot(df_pred['0'], kde=False, bins=bin_val)
# plt.xlim(-0.1,4.1)
plt.xticks(np.arange(0,5, 1))
plt.ylabel("Test Set: Bin Count", labelpad=10)
plt.xlabel('Classification')
plt.tight_layout()
plt.savefig('test_set_histogram.png')
```
